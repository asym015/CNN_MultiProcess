```python
import cv2

for i in range(1,387):   
          #NUMBER OF  picture -1 
          print(i)  
          img1 = cv2.imread(r"C:\Users\Hadil boussetta\Desktop\ok\filtrage sortie\1 ("+str(i)+").jpg")
          print(img1)
           #gray with rgb color 
          rgbgra=cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
          #convert the input image to grayscale
          bgrgra = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
          #  binary image
          ret,thresh = cv2.threshold(bgrgra,100,255,0)
          #convert the input image to HSV image
          bgrhsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
           #convert the input image to bright to lab 
          brightlab = cv2.cvtColor(img1, cv2.COLOR_BGR2LAB)
          cv2.imshow("Original Image",img1)
          cv2.imshow("rgb To Gray",rgbgra)
          cv2.imshow("BGR To Gray",bgrgra)
          cv2.imshow("BGR To HSV", bgrhsv)
          cv2.imshow("Bright To Labl", brightlab)
          cv2.imwrite("test0("+str(i)+").jpg",rgbgra)
          cv2.imwrite("test1("+str(i)+").jpg",bgrgra)
          cv2.imwrite("test2("+str(i)+").jpg",bgrhsv)
          
          cv2.imwrite("test3("+str(i)+").jpg",brightlab)
          cv2.imwrite("test4("+str(i)+").jpg", thresh) 
```
